```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

students_data = pd.read_csv('student-mat.csv', sep=";")
students_data = students_data[["G1","G2","studytime","failures","absences","G3"]]

X = students_data.drop(columns=['G3'])
y = students_data['G3']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train.values, y_train)

predictions = lr.predict(X_test.values)

plt.figure(figsize=(10, 7))
plt.scatter(y_test.values, predictions, alpha=0.5)
plt.plot([min(y_test.values), max(y_test.values)], [min(y_test.values), max(y_test.values)], color='red', linestyle='-', linewidth=1)
plt.title('Actual vs Predicted')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```
